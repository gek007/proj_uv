```python
from abc import ABC, abstractmethod


class Shape(ABC):   

    @abstractmethod
    def area(self):
        pass

    @abstractmethod 
    def perimeter(self):
        pass



class Rectangle(Shape):

    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)
    


print(type(Rectangle))    


class A:
    def display(self):
        print("Class A")



print(type(A))



class B:

    instance = None

    def __new__(cls):

        if cls.instance is None:
            cls.instance = super().__new__(cls)
        return cls.instance
    

a = B()
b = B()

print(a is b)


def singletom(cls):
    instances = {}

    def get_instance(*args, **kwargs):
        if cls not in instances:
            instances[cls] = cls(*args, **kwargs)
        return instances[cls]
    
    return get_instance

@singletom
class DataBase:
    pass


z = DataBase()
y = DataBase()
print(z is y)


def log(f: int) -> int:
    print(f"Function {f.__name__} is called")
    return f



from asyncio import threads
import time
import threading



def worker():
    print("Worker thread is running")
    time.sleep(2)
    print("Worker thread is done")



threads = [threading.Thread(target=worker) for _ in range(3)]    

for t in threads:   
    t.start()

for t in threads:
    t.join()    


from multiprocessing import Process


def worker_process():   
    print("Worker process is running")
    time.sleep(2)
    print("Worker process is done")


procs = [Process(target=worker_process) for _ in range(3)]
for p in procs:
    p.start()

for p in procs:
    p.join()


import asyncio


async def async_worker():
    print("Async worker is running")
    await asyncio.sleep(2)
    print("Async worker is done")


async def main():
    task1 = asyncio.create_task(async_worker())
    task2 = asyncio.create_task(async_worker())

    await task1
    await task2



    


i = 0 
for _ in range(3):
    print(f"Starting async event loop {i}")
    i += 1






```
